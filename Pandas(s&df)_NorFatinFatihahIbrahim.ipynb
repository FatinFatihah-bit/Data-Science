```python
# numpy array
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
# pandas series
# A series is 1D labeled array

srs = pd.Series([10,20,30,40])
print(type(srs))
print(srs)
print(srs.shape)



# in numpy, only numbers
# in series, numbers + labels (index)
```

```python
info = pd.Series(['Ali', 29, 9800, 'JB'], index=['name', 'age', 'salary', 'city'])
print(info)
print(info['salary'])
print(info['name'])        # indexing
print(info.get('city'))    # getting data values using get()
# print(info[2]) ==> dia akan bagi warning..so use .iloc instead
print(info.iloc[2])        # iloc --> Index Location

# Series is like a dictionary + array combo
```

```python
# values
print(info.values)
```

```python
# indexing slicing and using loops
s = pd.Series([10,20,30,40,50,60], index=['val1','val2','val3','val4','val5','val6'])
print(s)

# 10  20  30  40  50  60
# 0   1   2   3   4   5
# -6 -5  -4  -3  -2  -1
# v1 v2  v3  v4  v5  v6

# Indexing in Series can be done by using [] or .iloc[] or .loc[]
print('\nIndexing In Series')
print(s.iloc[0])     # for index numbers
print(s.loc['val5'])   # Loc used for Labels
print(s['val2'])       # [] used for labels
print(s.iloc[-1])
```

```python
# slicing in series
print('\nSlicing In Series')
print(s[1:4])       # 4 will be excluded --> stopIndex
print(s['v1':'v5']) # if perform slicing using Labels, the end Label in included

# 10, 40, 60
print(s[['val1','val4','val6']]) # for specific information s[['','','']]
print(s[['val1','val4','val6']].values) #nk values saja
```

```python
# conditional selection
print('\nConditional Selection')
print(s[s%3==0])
```

```python
# using for loop
for i in s:
  print(i)
```

```python
for k,v in s.items():
  print(k,v)
```

```python
for i in enumerate(s):
  print(i)
```
