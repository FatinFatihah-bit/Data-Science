```python
# import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data
data = pd.read_csv("https://raw.githubusercontent.com/realpython/python-data-cleaning/refs/heads/master/Datasets/BL-Flickr-Images-Book.csv")
data
```

```python
# basic info
data.info()
```

```python
# drop unnecessary columns

to_drop = ['Edition Statement',
            'Corporate Author',
           'Corporate Contributors',
            'Former owner',
            'Engraver',
            'Contributors',
           'Issuance type',
            'Shelfmarks']

data.drop(to_drop, inplace=True, axis=1)
```

```python
data.head()
```

```python
data.head(30)
```

```python
# check for missing values
data.isnull().sum()
```

```python
# fill the missing values

data['Author'].fillna(value='UnknownAuthor',inplace=True)
data['Publisher'].fillna(value='PublisherNotKnown',inplace=True)
```

```python
data.head(30)
```

```python
data.loc[206]
```

```python
# remove spaces and lowercase all column names
data.columns = data.columns.str.strip().str.lower()
data.columns
```

```python
# extract only 4-digit year using regex
data['date of publication'] = data['date of publication'].str.extract(r'(\d{4})', expand=False)

# Convert to numeric
data['date of publication'] = pd.to_numeric(data['date of publication'], errors='coerce')
```

```python
# fix invalid year values
data = data[(data['date of publication'] >= 1000) | (data['date of publication'].isnull())]
```

```python
# Fill Missing Values (Date of Publication) By using mean

mean_val_marks = data['date of publication'].mean()
data['date of publication'].fillna(value=mean_val_marks,inplace=True)
data.head(30)
```

```python
data['date of publication'] = data['date of publication'].astype(str).str.replace('.0', '', regex=False)
data.head(2)
```

```python
# check duplicates
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
# final check
data.info()
data.isnull().sum()
data.head(3)
```
